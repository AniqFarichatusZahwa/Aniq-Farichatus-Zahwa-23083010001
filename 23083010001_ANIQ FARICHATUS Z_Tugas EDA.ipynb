```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r'C:\Users\ANIQ FARICHATUS\Downloads\archive (5)\googleplaystore_user_reviews.csv')
print(data.head())
```

### Bar Plot Distribusi Sentimen

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=data)
plt.title('Distribusi Sentimen Aplikasi')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.show()
```

### Distribusi polaritas sentimen

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Sentiment_Polarity'], bins=20, kde=True)
plt.title('Distribusi Polaritas Sentimen')
plt.xlabel('Polaritas Sentimen')
plt.ylabel('Frekuensi')
plt.show()
```

### Distribusi subjektivitas sentimen

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Sentiment_Subjectivity'], bins=20, kde=True)
plt.title('Distribusi Subjektivitas Sentimen')
plt.xlabel('Subjektivitas Sentimen')
plt.ylabel('Frekuensi')
plt.show()
```

### Box plot polaritas sentimen berdasarkan sentimen (positif, negatif, netral)

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sentiment', y='Sentiment_Polarity', data=data)
plt.title('Polaritas Sentimen berdasarkan Kategori Sentimen')
plt.xlabel('Kategori Sentimen')
plt.ylabel('Polaritas Sentimen')
plt.show()
```
